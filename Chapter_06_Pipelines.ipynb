# Chapter 6 - Pipelines

## Ringkasan Chapter
Pada chapter ini dipelajari konsep utama terkait Pipelines dalam Machine Learning menggunakan Python dan scikit-learn.

## Teori
Pipeline menggabungkan preprocessing dan model.

```
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.datasets import load_iris
iris=load_iris()
```

```
pipe=Pipeline([('scaler',StandardScaler()),('svm',SVC())])
pipe.fit(iris.data,iris.target)
print(pipe.score(iris.data,iris.target))
```

## Penjelasan
Pipeline mempermudah workflow ML.

## Kesimpulan
Model telah berhasil dijalankan dan menunjukkan bagaimana konsep Machine Learning bekerja secara praktis. 
Dari eksperimen ini dapat disimpulkan bahwa pemilihan model dan preprocessing sangat mempengaruhi hasil akhir.
