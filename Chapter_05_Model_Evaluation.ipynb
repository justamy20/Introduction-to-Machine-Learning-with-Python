# Chapter 5 - Model Evaluation

## Ringkasan Chapter
Pada chapter ini dipelajari konsep utama terkait Evaluasi Model dalam Machine Learning menggunakan Python dan scikit-learn.

## Teori
Evaluasi menggunakan cross-validation dan metrics.

```
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_iris
iris=load_iris()
```

```
model=LogisticRegression(max_iter=200)
scores=cross_val_score(model,iris.data,iris.target,cv=5)
print(scores)
```

## Penjelasan
Cross-validation membagi data untuk evaluasi lebih akurat.

## Kesimpulan
Model telah berhasil dijalankan dan menunjukkan bagaimana konsep Machine Learning bekerja secara praktis. 
Dari eksperimen ini dapat disimpulkan bahwa pemilihan model dan preprocessing sangat mempengaruhi hasil akhir.
