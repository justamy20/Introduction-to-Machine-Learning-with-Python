# Chapter 7 - Text Data

## Ringkasan Chapter
Pada chapter ini dipelajari konsep utama terkait Text Data dalam Machine Learning menggunakan Python dan scikit-learn.

## Teori
Text harus diubah menjadi numerik menggunakan vectorization.

```
from sklearn.feature_extraction.text import CountVectorizer
texts=['machine learning','learning AI']
```

```
vec=CountVectorizer()
X=vec.fit_transform(texts)
print(vec.get_feature_names_out())
print(X.toarray())
```

## Penjelasan
CountVectorizer mengubah teks menjadi fitur numerik.

## Kesimpulan
Model telah berhasil dijalankan dan menunjukkan bagaimana konsep Machine Learning bekerja secara praktis. 
Dari eksperimen ini dapat disimpulkan bahwa pemilihan model dan preprocessing sangat mempengaruhi hasil akhir.
