# Chapter 3 - Unsupervised Learning

## Ringkasan Chapter
Pada chapter ini dipelajari konsep utama terkait Unsupervised Learning dalam Machine Learning menggunakan Python dan scikit-learn.

## Teori
Digunakan untuk menemukan pola tanpa label seperti clustering.

```
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
iris=load_iris()
```

```
kmeans=KMeans(n_clusters=3,random_state=0)
kmeans.fit(iris.data)
print(kmeans.labels_)
```

## Penjelasan
KMeans mengelompokkan data menjadi beberapa cluster.
