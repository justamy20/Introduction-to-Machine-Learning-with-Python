# Chapter 2 - Supervised Learning

## Ringkasan Chapter
Pada chapter ini dipelajari konsep utama terkait Supervised Learning dalam Machine Learning menggunakan Python dan scikit-learn.

## Teori
Supervised learning menggunakan data berlabel untuk klasifikasi dan regresi.

```
from sklearn.datasets import load_breast_cancer
data=load_breast_cancer()
```

```
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(data.data,data.target,random_state=0)
```

```
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(max_iter=5000)
model.fit(X_train,y_train)
print(model.score(X_test,y_test))
```

## Penjelasan
Logistic Regression digunakan untuk klasifikasi dua kelas.

## Kesimpulan
Model telah berhasil dijalankan dan menunjukkan bagaimana konsep Machine Learning bekerja secara praktis. 
Dari eksperimen ini dapat disimpulkan bahwa pemilihan model dan preprocessing sangat mempengaruhi hasil akhir.
