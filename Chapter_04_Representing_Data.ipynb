# Chapter 4 - Feature Engineering

## Ringkasan Chapter
Pada chapter ini dipelajari konsep utama terkait Feature Engineering dalam Machine Learning menggunakan Python dan scikit-learn.

## Teori
Feature engineering meningkatkan kualitas data.

```
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris
iris=load_iris()
```

```
scaler=StandardScaler()
X_scaled=scaler.fit_transform(iris.data)
print(X_scaled[:3])
```

## Penjelasan
Scaling membantu model bekerja lebih stabil.
